```python
import pandas as pd

from sklearn.linear_model import LogisticRegression

import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('student.csv')
```

```python
print(data.head())
print(data.isnull().sum())
print(data.describe)
```

```python
plt.scatter(data['study_hours'],data['pass'])
plt.show()
```

```python
features = data[['study_hours','attendance_percent','previous_score']]

result = data['pass']
```

```python
model = LogisticRegression()

model.fit(features,result)
```

```python
study_hours = float(input("Enter Study Hours : "))

attendance = float(input("Enter Attendace : "))

previous_score = float(input("Enter Privious Score : "))

final_result = model.predict([[study_hours,attendance,previous_score]])

if final_result > 0.7:
    print("According to details Student are likely to Pass")
    
else:
    print("Student are Likely to Fail in Exam")
```

```python
import pickle as pkl
pkl.dump(model,open('LogisticModel.pkl','wb'))
```
